```python
emotion_labels = ['anger', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'neutral']
isbn = []
emotion_scores = {label: [] for label in emotion_labels}

def split_sentences(text):
    return [
        s.strip()
        for s in text.split(".")
        if len(s.strip()) > 10
    ]

def calculate_max_emotion_scores(predictions):
    per_emotion_scores = {label: [] for label in emotion_labels}

    for sentence_pred in predictions:
        sorted_preds = sorted(sentence_pred, key=lambda x: x['label'])
        for idx, label in enumerate(emotion_labels):
            per_emotion_scores[label].append(sorted_preds[idx]["score"])

    return {label: np.max(scores) for label, scores in per_emotion_scores.items()}

from tqdm import tqdm

for i in tqdm(range(len(books))):
    text = books["description"][i]
    if not isinstance(text, str):
        continue

    isbn.append(books["isbn13"][i])

    sentences = split_sentences(text)
    predictions = classifier(sentences)

    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])

emotions_of = pd.DataFrame(emotion_scores)
emotions_of["isbn13"] = isbn

books = pd.merge(books, emotions_of, on="isbn13")
books.to_csv("books_with_emotions.csv", index=False)
```
